**Ex.1. A teacher wants to estimate the mean height of all 400 students at her school. She takes a simple random sample of 30 students and finds the sample mean height is 63 inches with a standard deviation of 1.5 inches. Construct a 90% confidence interval for the population
mean height. Round to two decimal places**

```python
import numpy as np
import scipy.stats as stats
# Given Data
n = 30
mean = 63
std_dev = 1.5
confidence = 0.90
# Sample size
# Sample mean (inches)
# Sample standard deviation (inches)
# Confidence level
# Calculate Standard Error (SE = std_dev / sqrt(n))
standard_error = std_dev / np.sqrt(n)
# Degrees of freedom (n-1)
dof = n - 1
# Calculate the Confidence Interval using t-distribution
ci = stats.t.interval(confidence, dof, loc=mean, scale=standard_error)
# Round to two decimal places
lower_bound = round(ci[0], 2)
upper_bound = round(ci[1], 2)
print(f"90% Confidence Interval: ({lower_bound}, {upper_bound}) inches")
```

**Ex.2. For a class project, a political science student at a large university wants to estimate the percent of students who are registered
voters. He surveys 500 students and finds that 300 are registered voters. Compute a 90% confidence interval for the true percent of
students who are registered voters, and interpret the confidence interval**

```python
import math
from scipy import stats
n = 500
x = 300
p_hat = x / n
confidence_level = 0.90
# Calculate Z-score
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
# Calculate Standard Error and Margin of Error
standard_error = math.sqrt((p_hat * (1 - p_hat)) / n)
margin_of_error = z_score * standard_error
# Calculate Bounds
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error
print(f"90% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```
