**Solve the example for One sample t-test:  
Lab Experiment 1 using python
Ex- Given a sample with mean 300 with standard deviation of 18.5 and sample size of 20.
Can it be assumed to be drawn from a population with mean 310 at 5% Los.
S olution:  
Here:  Sample size (n) = 20, x̄ = sample mean = 300, μ = population mean = 310, sample
standard deviation = s = 18.5**

```python
# One Sample t-test

from scipy import stats
import numpy as np

# Given data
sample_mean = 300
population_mean = 310
sample_std_dev = 18.5
sample_size = 20

# Calculate standard error
standard_error = sample_std_dev / np.sqrt(sample_size)

# Calculate t-statistic
t_statistic = (sample_mean - population_mean) / standard_error

# Degrees of freedom
df = sample_size - 1

# Calculate p-value (Two-Tailed Test)
p_value = stats.t.sf(abs(t_statistic), df) * 2

# Print results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Decision Making
alpha = 0.05

if p_value > alpha:
    print("Fail to Reject H0")
    print("Sample is drawn from population with mean 310")
else:
    print("Reject H0")
    print("Sample is NOT drawn from population with mean 310")
```

**Lab Experiment 3 using python:  
Use Random seed to generate two samples Data1 and Data2, deriving t-statistics and p-value from the data. Interpret your result at 5% los**

```python
from numpy.random import seed
from numpy.random import randn
from scipy.stats import ttest_ind
#seed the random number generator
seed(1)
#generate two independent samples
data1 = 5 * randn(100) + 50
data2 = 5 * randn(100) + 51
#compare samples
stat, p = ttest_ind(data1, data2)
print('Statistics=%.3f, p=%.3f' % (stat, p))
#interpret
alpha = 0.05
if p > alpha: #if p>0.05 then it accepts H1 and reject H0
	print('Same distributions (fail to reject H0)')
else:
	print('Different distributions (reject H0)')
```

**Lab Experiment 4 using python:
i. Check the goodness of fit for the following data mapping number of hours a student
studies v/s the expected number of hours at 5% los.
Observed Data (in hours) 8 6 10 7 8 11 9
Expected Data  (in hours) 9 8 11 8 10 7 6**

```python
# Chi-Squared Goodness of Fit Test

from scipy.stats import chisquare

# Observed data
observed = [8, 6, 10, 7, 8, 11, 9]

# Expected data
expected = [9, 8, 11, 8, 10, 7, 6]

# Perform Chi-Squared test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Print results
print("Chi-Square Statistic =", chi_stat)
print("P-value =", p_value)

# Level of significance
alpha = 0.05

# Interpretation
if p_value > alpha:
    print("Fail to Reject H0")
    print("Observed data fits the expected data")
else:
    print("Reject H0")
    print("Observed data does NOT fit the expected data")
```

**ii. Check the goodness of fit for the following data mapping number of hours a student
