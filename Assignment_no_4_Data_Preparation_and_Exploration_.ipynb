```python
import pandas as pd
import numpy as np
import io

# The dataset provided in the problem description
data_str = """
Chicago IL
San Antonio TX 22259
 TX 77343
 TX 4927
San Diego CA 16150
Philadelphi a PA
New York  25235
San Antonio 73868
Los Angeles CA 24607
San Diego CA 39289
"""

# Custom parsing logic for the unstructured data
records = []
for line in data_str.strip().split('\n'):
    parts = line.strip().split()
    city_name = []
    state_code = np.nan
    zip_code = np.nan

    if not parts:
        continue

    # Try to extract zip code (numeric) from the end
    if parts and parts[-1].isdigit():
        zip_code = parts[-1]
        parts.pop()

    # Try to extract state code (2 uppercase letters) from the end of remaining parts
    if parts and len(parts[-1]) == 2 and parts[-1].isupper():
        state_code = parts[-1]
        parts.pop()

    # The remaining parts form the city name
    city_name = " ".join(parts) if parts else np.nan

    records.append({'City': city_name, 'State': state_code, 'Zip': zip_code})

df_out = pd.DataFrame(records)

# Correct specific typo if present
df_out.loc[df_out['City'] == 'Philadelphi a', 'City'] = 'Philadelphia'

print("Original DataFrame (after custom parsing):\n", df_out)
print("\nMissing values before handling:\n", df_out.isnull().sum())
```

```python
# Handle missing values
# For categorical columns (City, State, Zip), filling with 'Unknown' is a common strategy.

df_out['City'].fillna('Unknown', inplace=True)
df_out['State'].fillna('Unknown', inplace=True)
df_out['Zip'].fillna('Unknown', inplace=True)

print("\nDataFrame after handling missing values:\n", df_out)
print("\nMissing values after handling:\n", df_out.isnull().sum())
```
