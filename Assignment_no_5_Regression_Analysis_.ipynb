```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data for Example 1
x = np.array([2, 3, 5, 7, 9, 10])
y = np.array([2, 5, 8, 10, 12, 14])

print("Example 1 Data Loaded:")
print(f"x: {x}")
print(f"y: {y}")
```

**Lab Experiment 1: Fit the data given in Example 1 to a regression line using python
program.**

```python
# Lab Experiment 1: Fit the data to a regression line

# Calculate coefficients for a linear regression line (y = mx + c)
m, c = np.polyfit(x, y, 1)

# Generate y-values for the regression line
y_reg = m * x + c

# Plotting the original data and the regression line
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, label='Original Data', s=100, color='blue')
sns.lineplot(x=x, y=y_reg, color='red', label=f'Regression Line (y = {m:.2f}x + {c:.2f})')
plt.title('Lab Experiment 1: Linear Regression Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

**Lab Experiment 2: Fit the data given in Example 1 to a one degree polynomial using a
python program (Hint: use a polyfit function)**

```python
# Lab Experiment 2: Fit the data to a one-degree polynomial using polyfit

# Using np.polyfit with degree 1 for a one-degree polynomial (linear fit)
coefficients = np.polyfit(x, y, 1)
m_poly, c_poly = coefficients[0], coefficients[1]

# Generate y-values for the polynomial line
y_poly = m_poly * x + c_poly

# Plotting the original data and the one-degree polynomial fit
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y, label='Original Data', s=100, color='green')
sns.lineplot(x=x, y=y_poly, color='purple', label=f'1-Degree Polynomial Fit (y = {m_poly:.2f}x + {c_poly:.2f})')
plt.title('Lab Experiment 2: One-Degree Polynomial Fit')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Data for Lab Experiment 3 (Temperature vs. Pressure)
temperature = np.array([0, 20, 40, 60, 80, 100])
pressure = np.array([0.0002, 0.0012, 0.006, 0.03, 0.09, 0.27])

print("Lab Experiment 3 Data Loaded:")
print(f"Temperature (x): {temperature}")
print(f"Pressure (y): {pressure}")
```

```python
# Lab Experiment 3: Fit second and third-degree regression polynomials

# Fit a second-degree polynomial
poly2_coeffs = np.polyfit(temperature, pressure, 2)
poly2_func = np.poly1d(poly2_coeffs)

# Fit a third-degree polynomial
poly3_coeffs = np.polyfit(temperature, pressure, 3)
poly3_func = np.poly1d(poly3_coeffs)

# Generate points for the fitted curves
temp_fit = np.linspace(temperature.min(), temperature.max(), 100)
pressure_poly2_fit = poly2_func(temp_fit)
pressure_poly3_fit = poly3_func(temp_fit)

# Plotting the original data and the polynomial fits
plt.figure(figsize=(10, 7))
sns.scatterplot(x=temperature, y=pressure, label='Original Data', s=100, color='blue', zorder=5)
sns.lineplot(x=temp_fit, y=pressure_poly2_fit, color='red', linestyle='--', label=f'2nd Degree Poly Fit (y = {poly2_coeffs[0]:.2e}x^2 + {poly2_coeffs[1]:.2e}x + {poly2_coeffs[2]:.2e})')
sns.lineplot(x=temp_fit, y=pressure_poly3_fit, color='green', linestyle='-', label=f'3rd Degree Poly Fit (y = {poly3_coeffs[0]:.2e}x^3 + {poly3_coeffs[1]:.2e}x^2 + {poly3_coeffs[2]:.2e}x + {poly3_coeffs[3]:.2e})')

plt.title('Lab Experiment 3: Second and Third Degree Polynomial Regression')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.legend()
plt.grid(True)
plt.show()
```
